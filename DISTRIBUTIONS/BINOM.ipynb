```python
from scipy import stats
p=0.5
n=1
dagilim=stats.binom(n,p)
yazi=dagilim.pmf(k=1) #1 denemede(n) 1 kez yazi(k) gelme olasiligi
print(yazi)
```

```python
# Tamami kusursuz/0 kusurlu(k=0) ve ya en fazla 2'si kusurlu(k<=2):
p=0.01 # kusurlu_urun/basari
n=10
dagilim=stats.binom(n,p).cdf(2)
print(dagilim)
```
