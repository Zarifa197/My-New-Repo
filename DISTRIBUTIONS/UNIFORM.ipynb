DISCRETE UNIFORM:

```python
from scipy import stats
n=6
dagilim=stats.randint(1,n+1) #ayrik uniform'u pythonda "randint" ile buluruz;son rakami almaz,o yuzden "n+1" yazdik.
olasilik=dagilim.pmf(k=5) #zarda 5 degerinin gelmesi
print(olasilik)
```

```python
beklenen_deger=dagilim.expect()
varyans=dagilim.var()
print(f"Beklenen deger {beklenen_deger},varyans {varyans}")
```

```python
olasilik=dagilim.cdf(2)
print(olasilik)
```

MONTE CARLO YONTEMI:

```python
import random
def zar_simulasyonu(deneme_sayisi):
    alti_sayisi=0
    for _ in range(deneme_sayisi):
        zar=random.randint(1,6)
        if zar==6:
            alti_sayisi+=1
    olasilik=alti_sayisi/deneme_sayisi
    return olasilik
sonuc=zar_simulasyonu(10000)
print(sonuc)
```

```python
def zar_simulasyonu(deneme_sayisi):
    alti_sayisi=0
    for _ in range(deneme_sayisi):
        zar=stats.randint(1,7).rvs()  #dagilimdan sayi cekmek icin "rvs" kullanilir
        if zar==6:
            alti_sayisi+=1
    olasilik=alti_sayisi/deneme_sayisi
    return olasilik
sonuc=zar_simulasyonu(10000)
print(sonuc)
```

CONTINUOUS UNIFORM:

```python
a=0
b=15
dagilim=stats.uniform(a,b) #surekli oldugu icin artik uniform cagiriyoruz
beklenen_deger=dagilim.expect()
print(beklenen_deger)
```

```python
olasilik=dagilim.cdf(12.5)
print(olasilik)
```
