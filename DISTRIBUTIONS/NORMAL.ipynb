```python
from scipy import stats
ortalama=500
standart_sapma=10
dagilim=stats.norm(ortalama,standart_sapma)
olasilik=dagilim.cdf(518)
print(olasilik*100)
```

```python
import numpy as np
ortalama=3000
varyans=90000
dagilim=stats.norm(ortalama,np.sqrt(varyans))
olasilik=dagilim.cdf(3500)
print((1-olasilik)*100)
```

```python
ortalama=4300
varyans=562500
dagilim=stats.norm(ortalama,np.sqrt(varyans))
olasilik1=dagilim.cdf(4200)
olasilik2=dagilim.cdf(2500)
print((olasilik1-olasilik2)*100)
```

```python
olasilik=dagilim.cdf(3000)
olasilik2=dagilim.sf(3000)  #survival function/sf(x)=1-cdf(x);yani burada 3000'den buyuk olma olasiligini direkt bulduk
print((1-olasilik)*100,olasilik2*100)
```

NORMALLIK VARSAYIMI:

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
veri=stats.norm.rvs(loc=38,size=100)
print(stats.kurtosis(veri))  #skewness ve kurtosis sinirlari -1.5,1.5 arasindaysa,normallik varsayimi odenir
print(stats.skew(veri))
sns.displot(veri,kde=True)
plt.show()
```

```python
stats.probplot(veri,dist="norm",plot=plt)
plt.show()
```

```python
ks=stats.kstest(veri,cdf="norm",args=(veri.mean(),veri.std()))
print(f"{ks.pvalue:5f}")  #alfa=%95 oldugu icin ve p>alfa oldugu icin H0 reddedilemez!Yani Kolmogorov-Smirnov testi sayesinde diyebiliriz ki,%95 guvenle bu veri normal dagilmis.
```

```python
sw=stats.shapiro(veri)
print(f"{sw.pvalue:5f}") 
```
