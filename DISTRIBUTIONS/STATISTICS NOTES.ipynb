ISTATISTIK'E GIRIS:

•Bilimsel yollarla toplanan verinin anlaşılır hale getirilmesinde kullanılan bi tekniktir istatistik.Veriler toplanır, sınıflandırılır, çözümlenir ve sonuçlar yorumlanır.

•Temel olarak 5 adet istatistiksel analiz süreci vardır. En basitden en zora doğru gider.

•İlk basamak betimsel istatistiksel analiz:sonuçlar tablolar,grafikler ve ya nicel(sayısal) şekilde raporlanır. 

•İkincisi çıkarımsal istatistiksel analiz:elimizdeki örneklem ile populasyonun özelliklerini elde etmeye çalışırız. Parametric(normal dağılıma uyan veri seti) ve non-parametric(uymayan veri seti) analizler var.

•Üçüncüsü fark istatistiği analizi:en az 2 farklı qrup arasında araşdırma yaparız ve fark varmı, yokmu ona bakarız,varsa,istatistiksel olarak anlamlımı? 

•Dördüncüsü ilişki analizi:olaylar arasında bi etkileşim, ilişki varmı, varsa ne anlamlımı? 

•Sonuncusu tahmin analizi:gelecekteki olayları tahmin ederiz. Mesela, hava durumu.


DEGISKENLER:

•DEĞİŞKENLER:Gözlemleri barındıran özellik yapılarıdır. 

•Gözlem-bir veri seti içindeki her bir değere verilen isimdir.

•Değişken tipleri :

1.Nicel / nitel(yapısına göre) :

*Nicel değişken:yalnizca sayisal değerlerden oluşur(boy,kilo)->quantitative

*Nitel değişken:sayilardan oluşmaz,sayili secim vardir(cinsiyet)->qualitative(kategorik)

2.Surekli / sureksiz(degerine gore):

*Surekli degisken:belirli bi aralikta sonsuza gider(boy,kilo)

*Sureksiz degiskenler:belirli bi gozleme karsilik gelen deger kisitlidir(cinsiyyet:1,0)

•PS:Nicel degiskenler hem surekli,hem sureksiz olurken;nitel degiskenler sadece sureksiz olur.

3.Bagimli / bagimsiz:

^Gunde kac sigara~kanser hucreleri sayisi

*Bagimsiz degisken:neden(gunde kac sigara)

*Bagimli degisken:sonuc(kanser hucreleri sayisi)

OLCEKLER:

•Olcekler:(en az bilgi verenden en cok bilgi verene dogru)

*Nominal:siniflandirma(daha degerli,daha az degerli meselesi yok)->nitel(ex:cinsiyyet)->mod,frekans bulunabilir.

*Ordinal:siralama(onem sirasi var)->nitel(ex:rutbeler)->mod,frekans,median,aciklik(range) bulunur.

*Interval(esik aralik olceyi)->metric:hem pozitif hem negatif degerlerin olceyi->nicel(ex:sicaklik degeri,zeka seviyesi,takvimler,katiliyorum/katilmiyorum gibi anket cevaplari)->mod,frekans,median,aciklik(range),ortalama,varyans,standart sapma bulunur.

*Ratio(oran olceyi)->metric:baslangic degeri 0 olan degerlerin olceyi->nicel(ex:aylik gelir,agirlik,hiz,uzunluk)->tum istatistiksel islemler yapilabilir.

PS:Nominalde kullana bilecegimiz ozelliyi ordinalde de kullana biliriz,ordinaldekini intervalde,hepsini de ratioda:cunku ratio hepsini kapsar

BETIMSEL ISTATISTIK'E GIRIS:

•Betimsel istatistik nedir?:istatistiksel analizin ilk kismidir;veri haqqinda ilk fikri verir

•2 grupa ayrilir:

*Merkezi egilim olculeri(ortalama(aritmetik),medyan,mod,kartil):verilerin konumu nedir;buyukmu/kucukmu,ortalamaya gor buyukmu/kucukmu?

*Merkezi dagilim olculeri(degisim araligi,standart sapma,varyans,carpiklik,basiklik):verilerin nasil bi alan yaratdigini gosterir

ARITMETIK ORTALAMA:

•Degerleri toplayip sayina boluyoruz.(Excel'de =AVERAGE/=ORTALAMA)

•Ayni bi degiskenin su anki ortalamasini,yarinki ortalamasi ile;ve ya baska bi degiskenin su anki ortalamasi ile kiyaslariz,gidisati kontrol ederiz.

MEDYAN:

•Her hangi iki sirket var diyelim:aylik gelirlerini karsilastiriyoruz;normal giderken bi anda sirketlerden biri nisan ayinda iflasin esiginden dondu ve geliri cok-cok dusdu->bu az gelire outlier(uc deger) denir ve bu ortalamayi kotu etkiler.Yani biz outlier'in oldugu durumlarda ortalamaya bakip saglikli karar veremeyiz.Bu zaman medyan devreye girer.

PS:Eger gozlemler arasinda bi uc deger yoksa,simmetrik;varsa,simmetrik olmayan dagilim sergilenir.


MOD(TEPE DEGERI):

•Veri setindeki degerlerin nerede toplandigini gosterir;frekansi en cok olani bulur.

•Frekans:her sayinin kac kec kullanildigini gosterir.

KARTIL(4'E BOLEN):

•Bi normal dagilim dusun:can egrisi.Bu egrinin tam ortasi mod(dagilimin zirvesi),median ve ortalamadir.

•Kartil bu grafigi 4 tane %25'lik parcalara boler.Toplam 3 tane kartil olur.Ikinci kartil ayni zamanda median,mod(dagilimin zirvesi) ve ortalamadir.

RANJ(DEGISIM ARALIGI):

•Seri(tum gozlemler) icerisindeki maksimum degerle minimum degerin farkidir.->veri sayisi cok oldugunda ranj anlamli olmayabilir.

•Ranjin buyuklugu degiskenin ayirtetme gucunun cok olmasini gosterir.Mesela,matematik sinavinda ranj 90,biyoloji sinavinda 50'yse,demek ki matematik sinavinda ogrencilerin puan farklari daha yuksek(ex:biri 100,biri 10 almis olabilir),bu yuzden de ogrencilerin seviyelerini test etmek icin matematik sinavi uygulansin diye biliriz

STANDART SAPMA:

•Standart sapmanin yuksek olmasi verilerin ortalamadan saptigini gosterir;outlier var->oynaklik var,o da riskin(volatilitenin) cok olmasi demek.

•Standart sapma'nin formulunde kare olmasinin sebebi su:X(i)-X(bar) hep 0 yapar.0'dan kurtulmak istiyoz.Sonra yeniden kok aliyoz ki,kare gitsin iste.

•Mesela standart sapmasi buyuk olan soru demek,az insanin cozdugu soru demek.

VARYANS:

•Standart sapmanin karesidir.

PS:Standart sapma ve varyans kucukse,qrafigin yuksekligi cok olur,yani dar olur;buyukse,o zaman genisligi buyuk olur,yanlara yayilir 

CARPIKLIK(SKEWNESS):

•Veride outlier varsa,grafikde carpiklik olusur.

•Ortalama>medyan'sa(positive outlier varsa),yani ortalama-medyan>0'sa,bu zaman pozitif yonde kuyruk olusur(saga carpik,positive skewness).

•Ortalama<medyan'sa(negative outlier varsa),yani ortalama-medyan<0'sa,bu zaman negatif yonde kuyruk olusur(sola carpik,negative skewness).

•Left-skew ya da right-skew da bi normal dagilim,aslinda.


KURTOSIS:

•Qrafigin yuksekligi(en basit halde).

•Kurtosis'in siniri 3 olarak kabul edilir.

•3'den buyukse,cok dik(leptokurtic) olur.

•3'se normal dagilim zaten(mezokurtic).

•3'den kucukse cok yayilir(platykurtic).


CIKARIMSAL ISTATISTIK:

•Anakutleden orneklem cekerek ona esasen anakutle hakkinda bilgi edinmeye calisiyoz.

PS1:Cikarimsal Istatistik:Inferential Statistics

PS2:Betimsel Istatistik:Descriptive Statistics

MERKEZI LIMIT TEOREMI(CLT):

•Uygun bi sekilde secilebilen orneklem yapisi anakutleye benzeyebilir.

•Populasyon(hangi dagilimda olursa olsun) icinden defalarca orneklem alirsak,bu orneklemlerin ortalamalari anakutle ortalamasi etrafinda normal dagilima yakinsar.(Orneklemin eleman sayisi 30 ve uzeri ise normal dagilim kabul edilir)

•Yani,diyelim ki,30000 elemanlik bi veri seti var.Ordan 2 elemanli orneklemi 1000 kez secelim ve her defasinda ortalama alalim(toplam 1000 tane ortalama olur).Ayni seyi sonra 30 elemanli orneklem cekerek yapalim.Elimizdeki ortalamalarin ortalamasi eleman sayisi buyudukce populasyonunkuna yakinsiycak.Ortalamalari grafiklersek,gorucez ki,o da normal dagilima yakinsiyor.

STANDART HATA:

•Bir anakutleden secilebilecek olasi orneklemlerin ortalamalarinin standart sapmasina standart hata denir.

•Standart hata ne kadar dusuk olursa,anakutle ortalamasina o kadar daha yakiniz demekdir.

Z-SCORE:

•Normal dagilim icinde verilerin kac standart sapma icerisinde yayildigini gosterir.

•Ortalamasi 0,standart sapmasi 1 olan dagilima standart normal dagilim denir.Bu halde z-score bulmak daha kolaydir.

•Dagilim icerisindeiki verilerin %68.2'si ortalamanin 1 standart sapma kadar uzaginda(-1,1 arasinda) kalir.

•Verilerin %95'i ortalamanin 2 standart sapma kadar uzaginda(-2,2 arasinda) kalir.

•Verilerin %99.7'si ortalamanin 3 standart sapma kadar uzaginda(-3,3 arasinda) kalir.

T DAGILIMI:

•T dağılımı ana kütle varyansının bilinmediyi,küçük örneklemlerde kullanılır(n<=30)

•Çünkü küçük örneklemlerde outlier tehlikesi daha büyük,yani 1 tane outlier oldumu biz bittik,ha bi de burda anakutle ortalamasi yok ya,o yuzden belirsizlik var,bu da kuyruklarin daha uzun olmasina yol acar(fat tail risk).T testi outlier'ları da hesaba katarak daha iyi sonuç verir.Ama df artdikca,yani orneklem buyudukce hata yapma olasiligimiz da azalir,bu yuzden gittikce T dagilimi normal dagilima yakinsar.
